```python
import pandas as pd
import gmplot
import mplleaflet
import matplotlib.pyplot as plt
import folium
```

```python
data = pd.read_csv("cl_apts.csv")
data
```

```python
coords = data[data["heart"] == True][["lat", "long"]]
std_coords = data[data["heart"] == False][["lat", "long"]]
```

```python
lat_mean, long_mean = coords.mean(axis=0)
std_lat_mean, std_long_mean = std_coords.mean(axis=0)
```

```python
map = folium.Map(location=[37.7569,-122.4145], zoom_start=13)
for coord in coords.iterrows():
    folium.CircleMarker(location=[coord[1].lat, coord[1].long],
                  color='red', 
                  radius=4,
                 ).add_to(map)
    
folium.Marker(location=[lat_mean, long_mean],
                  icon=folium.Icon(
                                  color='pink',
                                  icon='heart',    # icon code from above link
                                  prefix='glyphicon')
                 ).add_to(map)
map
```

```python
map = folium.Map(location=[37.7569,-122.4145], zoom_start=13)

folium.Marker(location=[lat_mean, long_mean],
                  icon=folium.Icon(
                                  color='pink',
                                  icon='heart',    # icon code from above link
                                  prefix='glyphicon')).add_to(map)
    
for coord in std_coords.iterrows():
    folium.CircleMarker(location=[coord[1].lat, coord[1].long],
                  color='gray', 
                  radius=4,
                 ).add_to(map)
    
for coord in coords.iterrows():
    folium.CircleMarker(location=[coord[1].lat, coord[1].long],
                  color='red', 
                  radius=4,
                 ).add_to(map)
    
map
```
